```python
try:
    score = eval(input("(100): "))
except:
    print("!")
else:
    addition = input("(y or n)? ")
    if addition == "y":
        score *= 1.05  # score=score * 1.05
        if score > 100:  # if  
            score = 100
    print(f"{round(score)}")
print("")
```
