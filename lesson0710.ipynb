#### [Exercise 1] 下列為換算成績等級的程式碼，換算的規則如下：

##### 90(含)~100分為「優」
##### 80(含)~89分為「甲」
##### 70(含)~79分為「乙」
##### 60(含)~69分為「丙」
##### 0(含)~59分為「丁」

```python
score=int(input("請輸入成績: "))
if score>=90:
    print("優")
elif score>=80:
    print("甲")
elif score>=70:
    print("乙")
elif score>=60:
    print("丙")
else:
    print("丁")
```

```python
try: 
    chinese:int=int(input("請輸入國文成績: "))
    math:int=int(input("請輸入數學成績: "))

    if chinese==100 and math==100:
        bonus=1000
    elif chinese==100 or math==100:
        bonus =500
    else:
        bonus =0

    print(f"獎金{bonus}")

except:
    print("輸入錯誤! 必須是整數")
```

### [Homework 3] 計算BMI,並輸出體重狀態

輸入:\
請輸入姓名:xxxx \
請輸入身高(cm):xxxx \
請輸入體重(kg):xxxx \

輸出:
bmi值請輸出至小數點2位數 \
xxx您好 \
身高是:xxx公分 \
體重是:xxxx公斤 \
BMI:xxx.xx \
您的體重:過重 \


```python
try:
    name: str = input("請輸入您的姓名: ")
    height: float = eval(input("請輸入您的身高(cm): "))
    weight: float = eval(input("請輸入您的體重(kg): "))
    print(f"{name}您好\n")
    print(f"身高是: {height}公分\n")
    print(f"體重是: {weight}公斤\n")
    BMI: float = weight/((height/100)**2)
    print(f"您的BMI值為: {round(BMI, ndigits=2)}\n")

    if BMI >=35:
        print("體位異常:重度肥胖")
    elif BMI>=30:
        print("體位異常:中度肥胖")
    elif BMI>=27:
        print("體位異常:輕度肥胖")
    elif BMI>=24:
        print("體位異常:過重")
    elif BMI>=18.5:
        print("健康體位")
    else:
        print("體重過輕")
except:
    print("輸入格式有誤")
```

### [Homework 4] 請輸入2數,求2數公因數

n = int(input('請輸入第1個數:')) \
m = int(input('請輸入第2個數:')) \
print("請輸入2數,求2數公因數:")


```python
n:int = int(input('請輸入第1個數:'))
m:int = int(input('請輸入第2個數:'))
if n<m:
    min=n
else:
    min=m
for i in range(1,min+1):
    if n%i==0 and m%i==0:

        print(i, end=' ')
```
