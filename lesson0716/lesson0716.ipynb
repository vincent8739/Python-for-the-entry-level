##### [Exercise 1] 定義一個函式 完全平方公式（x+y）^2=x^2+2xy+y^2

```python
def perfect_square (x, y):
    return x**2+2*x*y+y**2

perfect_square(10,20)
```

##### [Exercise 2] 定義一個函式, 輸入二個整數, 求其平均

```python
def average(n,m):
    avg=(n+m)/2
    return avg

n:int=int(input("please input one number:"))
m: int = int(input("please input the other number:"))
avg:float=average(n,m)
print(f"the average of two numbers {n},{m}: is {avg}")
```

##### [Exercise 3] 定義一個函式, 計算等差級數: 輸入數列的首項, 公差及項數 輸出末項及總和




```python
def initial_term(a1,d,n):
    an=a1+(n-1)*d
    return an

def sum(a1,d,n):
    sn = (a1 + an)*n/2
    return sn

a1:int=int(input("請輸入一個等差數列的首項(initial term): "))
d:int=int(input("請輸入一個等差數列的公差(common difference): "))
n:int=int(input("請輸入一個等差數列的項數(successive members): "))
an:int=initial_term(a1,d,n)
sn=sum(a1,d,n)

print(f"這個等差數列的末項(the n-term)是: {an}")
print(f"這個等差數列之合(sum)是: {sn}")
```
