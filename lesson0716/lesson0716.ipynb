##### [Exercise 2] , , 

```python
def average(n,m):
    avg=(n+m)/2
    return avg

n:int=int(input("please input one number:"))
m: int = int(input("please input the other number:"))
avg:float=average(n,m)
print(f"the average of two numbers {n},{m}: is {avg}")
```

##### [Exercise 3] , : ,  




```python
def initial_term(a1,d,n):
    an=a1+(n-1)*d
    return an

def sum(a1,d,n):
    sn = (a1 + an)*n/2
    return sn

a1:int=int(input("(initial term): "))
d:int=int(input("(common difference): "))
n:int=int(input("(successive members): "))
an:int=initial_term(a1,d,n)
sn=sum(a1,d,n)

print(f"(the n-term): {an}")
print(f"(sum): {sn}")
```
