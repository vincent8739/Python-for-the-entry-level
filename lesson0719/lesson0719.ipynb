```python
import pyinputplus as pyip

class Bmi():
    def __init__(self,na:str,he:int,we:int) -> None:
        self._name=na
        self._height=he
        self._weight=we
    
    @property
    def name(self) -> str:
        return self._name
    @property
    def height(self) -> float:
        return self._height
    @property
    def weight(self) -> float:
        return self._weight
  
    def get_bmi(self):
        BMI: float = self.weight/((self.height/100)**2)
        return BMI
    def get_status(self):
        BMI=self.get_bmi()
        if BMI >= 35:
            return ":"
        elif BMI >= 30:
            return ":"
        elif BMI >= 27:
            return ":"
        elif BMI >= 24:
            return ":"
        elif BMI >= 18.5:
            return ""
        else:
            return ""
    
def main():
        
    while True:
            
        try:
            selection:str=pyip.inputYesNo(prompt="BMI,'y''q':",yesVal="y",noVal="q") 

            if selection=='y':
                name: str = pyip.inputStr(": \n")
                height: float =pyip.inputFloat("(cm): \n",min=1)
                weight: float =pyip.inputFloat("(kg): \n", min=1)
                cal_BMI=Bmi(name,height,weight)
                BMI=cal_BMI.get_bmi()
                result=cal_BMI.get_status()
                print(f"{name}\n")
                print(f": {height}\n")
                print(f": {weight}\n")
                print(f"BMI: {round(BMI, ndigits=2)}, : {result} \n" )
                        
            elif selection=='q':
                print("")
                break

        except ValueError as ve:
            print(f": {ve}")
        except Exception as e:
            print(f": {e}")

if __name__ == "__main__":
    main()

        
```
