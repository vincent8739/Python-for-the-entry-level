```python
import pyinputplus as pyip

class Bmi():
    def __init__(self,na:str,he:int,we:int) -> None:
        self._name=na
        self._height=he
        self._weight=we
    
    @property
    def name(self) -> str:
        return self._name
    @property
    def height(self) -> float:
        return self._height
    @property
    def weight(self) -> float:
        return self._weight
  
    def get_bmi(self):
        BMI: float = self.weight/((self.height/100)**2)
        return BMI
    def get_status(self):
        BMI=self.get_bmi()
        if BMI >= 35:
            return "體位異常:重度肥胖"
        elif BMI >= 30:
            return "體位異常:中度肥胖"
        elif BMI >= 27:
            return "體位異常:輕度肥胖"
        elif BMI >= 24:
            return "體位異常:過重"
        elif BMI >= 18.5:
            return "健康體位"
        else:
            return "體重過輕"
    
def main():
        
    while True:
            
        try:
            selection:str=pyip.inputYesNo(prompt="請問要計算BMI,請按'y'或是離開請按'q':",yesVal="y",noVal="q") 

            if selection=='y':
                name: str = pyip.inputStr("請輸入您的姓名: \n")
                height: float =pyip.inputFloat("請輸入您的身高(cm): \n",min=1)
                weight: float =pyip.inputFloat("請輸入您的體重(kg): \n", min=1)
                cal_BMI=Bmi(name,height,weight)
                BMI=cal_BMI.get_bmi()
                result=cal_BMI.get_status()
                print(f"{name}您好\n")
                print(f"身高是: {height}公分\n")
                print(f"體重是: {weight}公斤\n")
                print(f"您的BMI值為: {round(BMI, ndigits=2)}, 結果為: {result} \n" )
                        
            elif selection=='q':
                print("感謝使用，再見！")
                break

        except ValueError as ve:
            print(f"輸入錯誤: {ve}")
        except Exception as e:
            print(f"發生未知錯誤: {e}")

if __name__ == "__main__":
    main()

        
```
