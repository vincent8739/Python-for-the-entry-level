### 三角函數基本概念



圓心: O 

半徑: OR = OC 

直角 90° : ∠OAC = ∠OBC = ∠OCD  

三角函數:

正弦 = sin θ = AC/OR = OB/OR = sine

餘弦 = cos θ = BC/OR = OA/OR = cosine

Θ = ∠COA = ∠OCB = ∠ACD = ∠OED

```python
import math
leg:float=eval(input("please input the right triangle of one leg: "))
hyp:float=eval(input("please input the right triangle of hypotenuse: "))
angle=math.degrees(math.asin(leg/hyp))
print(f"直角三角形其\n對邊長度為:{leg}\n斜邊長度為:{hyp}\n角度計算結果為:{round(angle,ndigits=2)}")
```

### 標準差
是一組數值自"平均值" 分散開來的程度的一種測量觀念。

一個較大的標準差，代表大部分的數值和其平均值之間"差異較大"；

一個較小的標準差，代表這些數值"較接近"平均值。

### 兩組數的集合{0, 5, 9, 14}和{5, 6, 8, 9} 求其平均值及標準差。







```python
import statistics
sum1:int=sum([0,5,9,14])
sum2:int=sum([5,6,8,9])
std1:float=statistics.stdev([0,5,9,14])
std2:float=statistics.stdev([5,6,8,9])
print(f"集合一[0,5,9,14]的總和為{sum1}, 其標準差為{std1}")
print(f"集合二[5,6,8,9] 的總和為{sum2}, 其標準差為{std2}")
```
