```python
import statistics
sum1:int=sum([0,5,9,14])
sum2:int=sum([5,6,8,9])
std1:float=statistics.stdev([0,5,9,14])
std2:float=statistics.stdev([5,6,8,9])
print(f"[0,5,9,14]{sum1}, {std1}")
print(f"[5,6,8,9] {sum2}, {std2}")
```
