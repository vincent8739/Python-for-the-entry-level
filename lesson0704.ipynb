# [Exercise 5] 請輸入一個半徑並計算出圓的面積 
### 圓面積=半徑x半徑x圓周率π



```python
import math
r:float=eval(input("please input a radius"))
area:float=r*r*math.pi
print(f"半徑為: {r} , 其圓面積是: {area}")
print(f"半徑為: {r} , 其圓面積是: {round(area,ndigits=4)}")
```

# [Homework#1] 請輸入身高,體重,計算BMI值

輸入:
請輸入姓名:xxxx
請輸入身高(cm):xxxx
請輸入體重(kg):xxxx

輸出:
bmi值請輸出至小數點2位數
xxx您好
身高是:xxx公分
體重是:xxxx公斤
BMI:xxx.xx



```python
name:str=input("請輸入您的姓名: ")
height:float=eval(input("請輸入您的身高(cm): ")) 
weight:float=eval(input("請輸入您的體重(kg): "))
BMI:float=weight/((height/100)**2)
print(f"{name}您好\n")
print(f"身高是: {height}公分\n")
print(f"體重是: {weight}公斤\n")
print(f"您的BMI值為: {round(BMI,ndigits=2)}")
```
