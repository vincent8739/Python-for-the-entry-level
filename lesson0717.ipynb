```python
class Person:
    def __init__(self,n: str):
        self.name=n
    def __repr__(self):
        return f"person, : {self.name}"

p1=Person(n="vincent lin")
p2=Person("")
print(p1)
print(p2)
```

```python
class Student(Person):
    def __init__(self, na:str,ch:int,en:int, ma:int ):
        super().__init__(n=na)
        self.chinese=ch
        self.english=en
        self.math=ma
    def __repr__(self):
        return "Student"
    def sum(self) -> int:
        total=self.chinese+self.english+self.math
        return total
    def average(self) -> float:
        total=self.sum()
        return total/3


s1=Student(na="",ch=96,en=91,ma=100)
print(s1.name)
print(s1.chinese)
print(s1.english)
print(s1.math)
print(s1)
print(s1.sum())
print(s1.average())

s2=Student("",93,96,90)
print(s2.name)
print(s2.chinese)
print(s2.english)
print(s2.math)
print(s2)
print(s2.sum())
print(s2.average())
```

### [Homework 8] BMI,
class Person\
__init__(self,name:str,height:int,weight:int)\
3attribute->name:str,height:int,weight:int\
2-> def getBmi() -> float, def get_status() -> str

```python
import pyinputplus as pyip

class Bmi():
    def __init__(self,na:str,he:int,we:int) -> None:
        self.name=na
        self.height=he
        self.weight=we
    def get_bmi(self):
        BMI: float = self.weight/((self.height/100)**2)
        return BMI
    def get_status(self):
        BMI=self.get_bmi()
        if BMI >= 35:
            return ":"
        elif BMI >= 30:
            return ":"
        elif BMI >= 27:
            return ":"
        elif BMI >= 24:
            return ":"
        elif BMI >= 18.5:
            return ""
        else:
            return ""
    
def main():
        
    while True:
            
        try:
            selection:str=pyip.inputYesNo(prompt="BMI,'y''q':",yesVal="y",noVal="q") 

            if selection=='y':
                name: str = pyip.inputStr(": ")
                height: float =pyip.inputFloat("(cm): ",min=1)
                weight: float =pyip.inputFloat("(kg):", min=1)
                cal_BMI=Bmi(name,height,weight)
                BMI=cal_BMI.get_bmi()
                result=cal_BMI.get_status()
                print(f"{name}\n")
                print(f": {height}\n")
                print(f": {weight}\n")
                print(f"BMI: {round(BMI, ndigits=2)}, : {result} \n" )
                        
            elif selection=='q':
                print("")
                break

        except ValueError as ve:
            print(f": {ve}")
        except Exception as e:
            print(f": {e}")

if __name__ == "__main__":
    main()

        
```
