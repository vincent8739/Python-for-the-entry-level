##### inputStr()- 接受一個字串。如果您基本上想要 Python 的input()或raw_input()，但需要 PyInputPlus 功能（例如白名單/黑名單、超時、限制等），請使用此選項。

```python
class Person:
    def __init__(self,n: str):
        self.name=n
    def __repr__(self):
        return f"我是person的實體, 我的名字是: {self.name}"

p1=Person(n="vincent lin")
p2=Person("林見恆")
print(p1)
print(p2)
```

```python
class Student(Person):
   pass

s1=Student("林同學")
print(s1.name)

```

```python
class Student(Person):
    def __init__(self, na:str,ch:int,en:int, ma:int ):
        super().__init__(n=na)
        self.chinese=ch
        self.english=en
        self.math=ma
    def __repr__(self):
        return "我是Student的實體"
    def sum(self) -> int:
        total=self.chinese+self.english+self.math
        return total
    def average(self) -> float:
        total=self.sum()
        return total/3


s1=Student(na="林同學",ch=96,en=91,ma=100)
print(s1.name)
print(s1.chinese)
print(s1.english)
print(s1.math)
print(s1)
print(s1.sum())
print(s1.average())

s2=Student("帥哥",93,96,90)
print(s2.name)
print(s2.chinese)
print(s2.english)
print(s2.math)
print(s2)
print(s2.sum())
print(s2.average())
```

### [Homework 8] 請將BMI的程式,使用物件導向的寫法
建立一個class Person\
一個__init__(self,name:str,height:int,weight:int)\
有3個attribute->name:str,height:int,weight:int\
有2個實體方法-> def getBmi() -> float, def get_status() -> str

```python
import pyinputplus as pyip

class Bmi():
    def __init__(self,na:str,he:int,we:int) -> None:
        self.name=na
        self.height=he
        self.weight=we
    def get_bmi(self):
        BMI: float = self.weight/((self.height/100)**2)
        return BMI
    def get_status(self):
        BMI=self.get_bmi()
        if BMI >= 35:
            return "體位異常:重度肥胖"
        elif BMI >= 30:
            return "體位異常:中度肥胖"
        elif BMI >= 27:
            return "體位異常:輕度肥胖"
        elif BMI >= 24:
            return "體位異常:過重"
        elif BMI >= 18.5:
            return "健康體位"
        else:
            return "體重過輕"
    
def main():
        
    while True:
            
        try:
            selection:str=pyip.inputYesNo(prompt="請問要計算BMI,請按'y'或是離開請按'q':",yesVal="y",noVal="q") 

            if selection=='y':
                name: str = pyip.inputStr("請輸入您的姓名: ")
                height: float =pyip.inputFloat("請輸入您的身高(cm): ",min=1)
                weight: float =pyip.inputFloat("請輸入您的體重(kg):", min=1)
                cal_BMI=Bmi(name,height,weight)
                BMI=cal_BMI.get_bmi()
                result=cal_BMI.get_status()
                print(f"{name}您好\n")
                print(f"身高是: {height}公分\n")
                print(f"體重是: {weight}公斤\n")
                print(f"您的BMI值為: {round(BMI, ndigits=2)}, 結果為: {result} \n" )
                        
            elif selection=='q':
                print("感謝使用，再見！")
                break

        except ValueError as ve:
            print(f"輸入錯誤: {ve}")
        except Exception as e:
            print(f"發生未知錯誤: {e}")

if __name__ == "__main__":
    main()

        
```
