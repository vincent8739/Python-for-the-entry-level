### [Homework 5] 求BMI(進階版) 
##### 執行完後,問使用者(離開[q],繼續[c])
##### 如果輸入格式錯誤,要求使用者重新輸入

```python
while True:
        
    try:
        selection:str=input("請問要計算BMI,請按'y'或是離開請按'q': ")
        if selection=='q':
            break
        elif selection=='y':

            name: str = input("請輸入您的姓名: ")
            height: float = eval(input("請輸入您的身高(cm): "))
            weight: float = eval(input("請輸入您的體重(kg): "))
            print(f"{name}您好\n")
            print(f"身高是: {height}公分\n")
            print(f"體重是: {weight}公斤\n")
            BMI: float = weight/((height/100)**2)
            print(f"您的BMI值為: {round(BMI, ndigits=2)}\n")

            if BMI >= 35:
                print("體位異常:重度肥胖")
            elif BMI >= 30:
                print("體位異常:中度肥胖")
            elif BMI >= 27:
                print("體位異常:輕度肥胖")
            elif BMI >= 24:
                print("體位異常:過重")
            elif BMI >= 18.5:
                print("健康體位")
            else:
                print("體重過輕")
    except:
        print("輸入格式有誤")
```

### [Exercise 4] 猜數字遊戲 數字範圍為: ,min~max  
### [Homework 6] 猜數字遊戲,請加入還要繼續嗎(y,n)?

```python
import random
min=1
max=100
count=0
lottery:int=random.randint(min,max)
print("------猜數字遊戲開始------\n")

while True:
    count+=1
    try:
        keyin=(input(f"請輸入一個數字,範圍從{min}到{max}, 若輸入[q]則離開遊戲"))
        if keyin=="q":
            break
        else:
            number:int=int(keyin)
    except:
        print("輸入格式錯誤!請從新輸入")
        continue

    if number<min or number>max:
        print("輸入數字超過範圍! 請從新輸入")
        continue
    
    if number>lottery:
        max = number-1
        print(f"輸入數字{number}太大, 請重新輸入")    
        print(f"目前猜了:{count}次")
    elif number<lottery:
        min=number+1
        print(f"輸入數字{number}太小, 請重新輸入")
        print(f"目前猜了:{count}次")
    else:
        number==lottery
        print(f"Bingo!數字是:{number}, 總共猜了:{count}次\n")
        break

print("------猜數字遊戲結束------")
```
