### [Homework 5] BMI() 
##### ,([q],[c])
##### ,

```python
while True:
        
    try:
        selection:str=input("BMI,'y''q': ")
        if selection=='q':
            break
        elif selection=='y':

            name: str = input(": ")
            height: float = eval(input("(cm): "))
            weight: float = eval(input("(kg): "))
            print(f"{name}\n")
            print(f": {height}\n")
            print(f": {weight}\n")
            BMI: float = weight/((height/100)**2)
            print(f"BMI: {round(BMI, ndigits=2)}\n")

            if BMI >= 35:
                print(":")
            elif BMI >= 30:
                print(":")
            elif BMI >= 27:
                print(":")
            elif BMI >= 24:
                print(":")
            elif BMI >= 18.5:
                print("")
            else:
                print("")
    except:
        print("")
```

### [Homework 6] ,(y,n)?

```python
import random
min=1
max=100
count=0
lottery:int=random.randint(min,max)
print("------------\n")

while True:
    count+=1
    try:
        keyin=(input(f",{min}{max}, [q]"))
        if keyin=="q":
            break
        else:
            number:int=int(keyin)
    except:
        print("!")
        continue

    if number<min or number>max:
        print("! ")
        continue
    
    if number>lottery:
        max = number-1
        print(f"{number}, ")    
        print(f":{count}")
    elif number<lottery:
        min=number+1
        print(f"{number}, ")
        print(f":{count}")
    else:
        number==lottery
        print(f"Bingo!:{number}, :{count}\n")
        break

print("------------")
```
